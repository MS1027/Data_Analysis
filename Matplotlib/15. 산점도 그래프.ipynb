# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
data = {'이름':['Bukayo Saka', 'Emile Smith Rowe', 'Martin Odegaard', 'Gabriel Martinelli', 'Edward Nketiah'],
       '득점':[11, 10, 7, 6, 5],
        '도움':[7, 2, 4, 5, 1],
       '출생년도':[2001, 2000, 1998, 2001, 1999]}
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['득점'], df['도움'])
plt.xlabel('득점')
plt.ylabel('도움')
```

```python
import numpy as np

sizes = np.random.rand(5) * 1000
sizes
```

```python
plt.scatter(df['득점'], df['도움'], s=sizes)
plt.xlabel('득점')
plt.ylabel('도움')
```

```python
sizes = (2005 - df['출생년도']) * 200
plt.scatter(df['득점'], df['도움'], s=sizes)
plt.xlabel('득점')
plt.ylabel('도움')
```

```python
plt.scatter(df['득점'], df['도움'], s=sizes, c=df['출생년도'], cmap='viridis', alpha = 0.5) # 출생년도별로 색상 구분
# colormap 변수로 정해진 colormap에서 랜덤으로 색상을 선정하는 것
plt.xlabel('득점')
plt.ylabel('도움')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['득점'], df['도움'], s=sizes, c=df['출생년도'], cmap='viridis', alpha = 0.5)
plt.xlabel('득점')
plt.ylabel('도움')
plt.colorbar(ticks=[1998, 1999, 2000, 2001], label='출생년도', shrink=0.5, orientation='horizontal')
```
