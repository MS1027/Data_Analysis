# 11. 누적 막대 그래프

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {'이름' : ['부카요 사카', '가브리엘 마르티넬리', '에디 은케티아', '에밀 스미스 로우'],
       '19/20' : [4, 6, 4, 0],
       '20/21' : [7, 2, 6, 4],
       '21/22' : [12, 6, 10, 11]}
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['이름'], df['19/20'], width=0.2)
plt.bar(df['이름'], df['20/21'], width=0.2)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['이름'], df['19/20'], width=0.2)
plt.bar(df['이름'], df['20/21'], width=0.2, bottom=df['19/20'])
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['이름'], df['19/20'], width=0.2)
plt.bar(df['이름'], df['20/21'], width=0.2, bottom=df['19/20'])
plt.bar(df['이름'], df['21/22'], width=0.2, bottom=df['19/20']+df['20/21'])
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df['이름'], df['19/20'], width=0.2, label='19/20')
plt.bar(df['이름'], df['20/21'], width=0.2, bottom=df['19/20'], label='20/21')
plt.bar(df['이름'], df['21/22'], width=0.2, bottom=df['19/20']+df['20/21'], label='21/22')

plt.xticks(rotation=30)
plt.legend()
```
