```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
data = {'이름' : ['부카요 사카', '가브리엘 마르티넬리', '에디 은케티아', '에밀 스미스 로우'],
       '19/20' : [4, 6, 4, 0],
       '20/21' : [7, 2, 6, 4],
       '21/22' : [12, 6, 10, 11]}
df = pd.DataFrame(data)
df
```

```python
index = np.arange(df.shape[1])
index
```

```python
w = 0.25
plt.bar(index-w, df['19/20'], width=w, label='19/20')
plt.bar(index, df['20/21'], width=w, label='20/21')
plt.bar(index+w, df['21/22'], width=w, label='21/22')
plt.legend()
```

```python
plt.figure(figsize=(10, 5))
plt.title("시즌별 득점 수")

w = 0.25
plt.bar(index-w, df['19/20'], width=w, label='19/20')
plt.bar(index, df['20/21'], width=w, label='20/21')
plt.bar(index+w, df['21/22'], width=w, label='21/22')
plt.legend()
plt.xticks(index, df['이름'])
plt.show()
```
