# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['MC', 'LIV', 'CHE', 'TOT', 'ARS', 'MU'] # 팀
values = [99, 94, 76, 69, 61, 57] # 리그 득점 수

plt.bar(labels, values)
```

```python
plt.bar(labels, values, color='g')
```

```python
colors = ['cyan', 'brown', 'blue', 'navy', 'red', 'orangered']

plt.bar(labels, values, color=colors, alpha=0.5) # 색상 개별 설정, 투명도 설정
```

```python
plt.bar(labels, values)
plt.ylim(50, 110) # y의 상/하한선 설정
```

```python
plt.bar(labels, values, width=0.5) # 막대 두께 (기본 값은 1)
```

```python
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45) # x축의 팀 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 득점 수 데이터 각도를 45도로 설정
```

```python
ticks = ['우승', '준우승', 'UCL', 'UCL', 'UEL', 'UEL']

plt.bar(labels, values)
plt.xticks(labels, ticks) # ticks 데이터를 x축 값으로 설정
```
