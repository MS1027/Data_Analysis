## 최근 3시즌 런던 3구단의 최종 승점

```python
seasons = ['19/20', '20/21', '21/22'] # 2019/20, 2020/21, 2021/22
ARS = [56, 61, 69] # Arsenal F.C.
CHE = [66, 67, 74] # Chelsea F.C.
TOT = [59, 62, 71] # Tottenham Hotspur F.C.

plt.plot(seasons, ARS, label='ARS', marker='v')
plt.plot(seasons, CHE, label='CHE', marker='o', ls='--')
plt.plot(seasons, TOT, label='TOT', marker='x', ls='-.')

plt.legend(ncol=3) # 3개의 열로 표시
```
