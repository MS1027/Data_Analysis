# 14. 원 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [11, 10, 7, 6, 5]
names = ['Bukayo Saka', 'Emile Smith Rowe', 'Martin Odegaard', 'Gabriel Martinelli', 'Edward Nketiah']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']

plt.pie(values, labels = names, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=[0.05]*5)
plt.show()
```

```python
wedgeprops={'width':0.6} # 파이의 두께
plt.pie(values, labels = names, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=[0.05]*5, wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth':5} # 외곽선 색상, 굵기
plt.pie(values, labels = names, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show() # explode와 달리 어긋나지 않음
```

```python
def custom_autopct(pct):
    return ('%.1f%%' % pct) if pct >= 20 else '' # 20% 이상인 경우에만 %를 표시
            # ':.1f'.format(pct)와 같음

plt.pie(values, labels = names, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show() # explode와 달리 어긋나지 않음
```
