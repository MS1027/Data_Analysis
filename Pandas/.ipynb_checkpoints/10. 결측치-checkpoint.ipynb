# 10. 결측치

비어 있는 데이터

```python
import pandas as pd
import numpy as np

df = pd.read_excel('10_ARS.xlsx', index_col='등번호')
df
```

## 데이터 채우기 fillna

```python
df.fillna('') # NaN 데이터를 빈 칸으로 채움
```

```python
df.fillna('MF') # NaN 데이터를 MF로 채움
```

```python
df
```

```python
df.fillna('MF', inplace=True)
```

```python
df
```

```python
df.loc[df['포지션'] == 'MF', '포지션'] = np.nan
```

```python
df.loc[df['국적'] == '브라질', '국적'] = np.nan
```

```python
df
```

```python
df['국적'].fillna('브라질', inplace=True)
```

```python
df
```

## 데이터 제외하기 dropna

```python
df.dropna() # 전체 데이터 중 NaN을 포함하는 데이터 삭제 (실제로 반영하기 위해서는 inplace 필요)
```

```python
df = pd.read_excel('10_ARS.xlsx', index_col='등번호')
df
```

- axis : index(기본값) or columns -> 행을 지울지, 열을 지울지  
- how : any(기본값) or all -> 1개라도 있을 경우, 전부 NaN일 경우  

```python
df.dropna(axis='index', how='any') # NaN이 하나라도 있는 row를 삭제
```

```python
df.dropna(axis='columns') # NaN이 하나라도 있는 Column을 삭제
```

```python
df['국적'] = np.nan
```

```python
df
```

```python
df.dropna(axis='columns', how='all') # 데이터 전체가 NaN인 '국적' Column 삭제
```
