# 14. 그룹화

동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('04_ARS.xlsx', index_col='등번호')
df
```

```python
df.groupby('포지션')
```

```python
df.groupby('포지션').get_group('FW')
```

```python
df.groupby('포지션').get_group('DF')
```

```python
df.groupby('포지션').mean() # 계산 가능한 데이터들의 평균값
```

```python
df.groupby('포지션').size() # 각 그룹의 크기
```

```python
df.groupby('포지션').size()['MF'] # 포지션으로 그룹화를 한 뒤에 MF에 해당하는 데이터의 수
```

```python
df.groupby('포지션')['키'].mean() # 포지션으로 그룹화를 한 뒤, 키의 평균 데이터
```

```python
df['세부 포지션'] = ['CF', 'WF', 'WF', 'CM', 'CM', 'DM', 'FB', 'CB', 'CB', 'FB', 'GK'] # 세부 포지션 column 추가
df
```

```python
df.groupby(['포지션', '세부 포지션']).mean() # 포지션별, 세부포지션별 평균 데이터
```

```python
df.groupby('세부 포지션').mean() # 세부 포지션별 평균 데이터
```

```python
df.groupby('포지션').mean().sort_values('키') # 포지션별 키 오름차순 정렬
```

```python
df.groupby('포지션')['국적'].value_counts() # 포지션으로 그룹화를 한 뒤에 각 포지션별 국적의 가짓수
"""
아래와 같은 결과
position = df.groupby('포지션')
position['국적'].value_counts()
"""
```

```python
df.groupby('포지션')['국적'].value_counts().loc['DF'] # 포지션으로 그룹화를 한 뒤에 DF에 대한 국적 별 선수 수
```

```python
df.groupby('포지션')['국적'].value_counts(normalize=True).loc['FW'] # FW의 국적 별 퍼센트
```
