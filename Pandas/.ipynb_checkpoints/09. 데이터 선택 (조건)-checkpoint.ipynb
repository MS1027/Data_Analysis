# 9. 데이터 선택 (조건)

조건에 해달하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('04_ARS.xlsx', index_col='등번호')
df
```

```python
df['키'] >= 180 # 선수들의 키가 180 이상인지 여부를 True/False
```

```python
myFilter = (df['키'] >= 180)
df[myFilter] # 필터를 만들어 조건에 참인 데이터만 가져옴
```

```python
df[~myFilter] # 필터를 역으로 적용 (키가 180 미만인 선수들)
```

```python
df[df['키'] >= 180] # 필터를 변수 지정 없이 사용
```

```python
df.loc[df['키'] >= 180, '포지션'] # 키가 180 이상인 선수들의 포지션
```

```python
df.loc[df['키'] >= 180, ['이름', '포지션', '키']] # 키가 180 이상인 선수들의 이름, 포지션, 키
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 180) & (df['포지션'] == 'MF')] # 키가 180 이상이고, 포지션이 MF인 선수 데이터
```

### | 또는

```python
df.loc[(df['출생년도'] < 1995 )|(df['출생년도'] > 2000)] # 1995년 이전 혹은 2000년 이후에 출생한 선수 데이터
```

### str 함수

```python
myFilter = df['이름'].str.startswith('가브리엘') # 이름이 가브리엘로 시작하는 선수
df[myFilter]
```

```python
myFilter = df['이름'].str.contains('카') # 이름에 '카'가 들어가는 선수
df[myFilter]
# NaN 데이터는 contains 함수의 결과값이 false가 아닌 NaN
# NaN이 포함된 필터를 사용하여 일반적으로 출력을 시도하면 오류가 발생함
# contains(키워드, na=True/False)로 사용하면 NaN을 True/False로 처리할 수 있음
```

```python
df[~myFilter] # 이름에 '카'가 들어가지 않는 선수
```

```python
nations = ['브라질', '잉글랜드']
myFilter = df['국적'].isin(nations) # 국적이 브라질이거나 잉글랜드인 사람
df[myFilter]
```
