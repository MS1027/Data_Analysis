# 7. 데이터 선택 (loc)

이름을 이용하여 원하는 row에서 원하는 col 선택

```python
import pandas as pd
df = pd.read_excel('04_ARS.xlsx', index_col='등번호')
df
```

```python
df.loc[9] # index 9에 해당하는 전체 데이터
```

```python
df.loc[12] # index 12에 해당하는 전체 데이터
```

```python
df.loc[12,'이름'] # index 12에 해당하는 '이름' 데이터
```

```python
df.loc[[11,7], '출생년도'] # index 11, 7에 해당하는 '출생년도' 데이터
```

```python
df.loc[[11, 7], ['이름', '국적']] # index 11, 7에 해당하는 '이름', '국적' 데이터
```

```python
df.loc[35:4, '이름':'국적'] # index 35부터 4까지, '이름'부터 '국적'까지의 데이터 (뒷부분 포함)
```
