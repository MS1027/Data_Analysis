## Series 확인

```python
df['키'] # column '키'에 해당하는 Series
```

```python
df['키'].describe()
```

```python
df['키'].min()
```

```python
df['키'].max()
```

```python
df['키'].nlargest(3) # 키가 큰 순서대로 3명의 데이터
```

```python
df['키'].mean() # 평균
```

```python
df['키'].sum() # 총합
```

```python
df['이름'].count() # 유효한 데이터 값의 갯수 (Null은 count되지 않음)
```

```python
df['국적'].unique() # 중복 제외
```

```python
df['국적'].nunique() # 중복을 제외한 결과의 수
```
