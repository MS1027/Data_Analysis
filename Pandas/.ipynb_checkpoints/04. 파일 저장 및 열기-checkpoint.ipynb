# 4. 파일 저장 및 열기

DataFrame 객체를 excel, csv, txt 등 형태의 파일로 만들기

```python
import pandas as pd
data = {
    '이름' : ['가브리엘 제수스', '가브리엘 마르티넬리', '부카요 사카', '그라니트 자카', '마르틴 외데고르', '토마스 파티', '올렉산드르 진첸코', '가브리엘 마갈량이스', '윌리엄 살리바', '벤 화이트', '아론 람스데일'],
    '포지션' : ['FW', 'FW', 'FW', 'MF', 'MF', 'MF', 'DF', 'DF', 'DF', 'DF', 'GK'],
    '국적' : ['브라질', '브라질', '잉글랜드', '스위스', '노르웨이', '가나', '우크라이나', '브라질', '프랑스', '잉글랜드', '잉글랜드'],
    '등번호' : [9, 11, 7, 34, 8, 5, 35, 6, 12, 4, 1],
    '키' : [175, 180, 178, 185, 176, 185, 175, 190, 193, 182, 188],
    '출생년도' : [1997, 2001, 2001, 1992, 1998, 1993, 1996, 1997, 2001, 1997, 1998]
}
df = pd.DataFrame(data)
df.set_index('등번호', inplace=True)
df
```

## 저장하기
### csv 파일로 저장

```python
df.to_csv('04_ARS.csv', encoding='utf-8-sig')
```

```python
df.to_csv('04_ARS.csv', encoding='utf-8-sig', index=False) # Index 제외하고 저장
```

### 텍스트(.txt) 파일로 저장

```python
df.to_csv('04_ARS.txt', sep='\t') # to_csv 함수를 사용하고, 확장자를 .txt로 해줌
```

### 엑셀 파일로 저장

```python
df.to_excel('04_ARS.xlsx')
```

***
## 열기

### csv 파일 열기

```python
df = pd.read_csv('04_ARS.csv')
df
```

```python
df = pd.read_csv('04_ARS.csv', skiprows=1) # 가장 위 1개의 줄 무시
df
```

```python
df = pd.read_csv('04_ARS.csv', skiprows=[1,3,5]) # 1,3,5번째 row는 제외 (0부터 시작)
df
```

```python
df = pd.read_csv('04_ARS.csv', nrows=4) # 지정된 개수만큼의 row만 가져옴
df
```

```python
df = pd.read_csv('04_ARS.csv', skiprows=2, nrows=4) # 처음 2 row 무시하고, 이후 4 row를 가져옴
df
```

### 텍스트(.txt) 파일 열기

```python
df = pd.read_csv('04_ARS.txt', sep='\t') # tab을 구분자로 설정하고 가져옴
df
```

```python
df = pd.read_csv('04_ARS.txt', sep='\t', index_col='등번호') # index를 지정하여 가져옴
"""
아래도 같은 결과
df = pd.read_csv('04_ARS.txt', sep='\t')
df.set_index('등번호', inplace=True)
"""
df
```

### 엑셀 파일 열기

```python
df = pd.read_excel('04_ARS.xlsx')
df
```

```python
df = pd.read_excel('04_ARS.xlsx', index_col='등번호')
df
```
