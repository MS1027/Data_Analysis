# 8. 데이터 선택 (iloc)

위치를 이용하여 원하는 row에서 원하는 col 선택

```python
import pandas as pd
df = pd.read_excel('04_ARS.xlsx', index_col='등번호')
df
```

```python
df.iloc[0] # 0번째 index에 해당하는 데이터
```

```python
df.iloc[4] # 4번째 index에 해당하는 데이터
```

```python
df.iloc[0:5] # 0 ~ 4번째 index에 해당하는 데이터
```

```python
df.iloc[3, 2] # 3번째 index의 2번째(0부터 시작) 데이터(국적)
```

```python
df.iloc[7, 4] # 7번째 index의 4번째 데이터 (출생년도)
```

```python
df.iloc[[0, 1], 0] # 0 ~ 1번째 index의 0번째 데이터(이름)
```

```python
df.iloc[[7, 8], [0, 2, 3]] # 7, 8번째 index의 0, 2, 3번째 데이터(이름, 국적, 키) 
```

```python
df.iloc[3:6, 0:3] # 3 ~ 5번째 index의 0 ~ 2번째 데이터(이름, 포지션, 국적)
```
