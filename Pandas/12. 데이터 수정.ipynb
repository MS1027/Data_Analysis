# 12. 데이터 수정

```python
import pandas as pd
import numpy as np
df = pd.read_excel('12_ARS.xlsx', index_col='등번호')
df
```

## Column 수정

```python
df['포지션'].replace({'Forward':'FW', 'Midfielder':'MF', 'Defender':'DF', 'Goalkeeper':'GK'}) # 실제 데이터에 반영 시 inplace 필요
```

## Column 추가

```python
df['이전 소속팀'] = ['Manchester City', 'Ituano FC', np.nan, 'Borussia M\'Gladbach', 'Real Madrid CF', 'Atletico Madrid', 'Manchester City', 'Lille', 'Saint-Etienne', 'Brighton & Hove Albion', 'Sheffield United']
df
```

```python
df['홈그로운'] = 'Non-Homegrown'
df
```

```python
df.loc[((df['출생년도'] >= 1999) | (df['국적'] == '잉글랜드')), '홈그로운'] = 'Homegrown'
df
```

## Column 삭제

```python
df.drop(columns=['이전 소속팀']) # 이전 소속팀 column 삭제 (inplace 필요)
```

```python
df.drop(columns=['이전 소속팀', '홈그로운']) # 이전 소속팀, 홈그로운 column 삭제
```

## Row 삭제

```python
df.drop(index=4) # 등번호(index)가 4인 선수(벤 화이트) 데이터 row 삭제
```

```python
myFilter = df['키'] < 180
df[myFilter]
```

```python
df[myFilter].index
```

```python
df.drop(index=df[myFilter].index)
```

## Row 추가

```python
df.loc[11] = ['에밀 스미스로우', 'Forward', '잉글랜드', 182, 2000, np.nan, 'Homegrown'] # 새로운 row 추가
df
```

## Cell 수정

```python
df.loc[11, '포지션'] = 'Midfielder'
df
```

```python
df.loc[11, ['이름','포지션']] = ['에밀 스미스 로우', 'Forward'] 
```

```python
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 맨 뒤에 있는 홈그로운 column을 가장 앞으로 이동 / [[리스트]]의 형태
df
```
